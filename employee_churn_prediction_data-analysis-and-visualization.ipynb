Qualitative observations based on looking at the histograms and the bar charts:

1. A major observation is that a large portion of employees leaving the company are actually young employees, who are at maximum in their early 30’s. In lower age groups, the attrition is relatively high. Age.

2. Churn rate is negatively correlated to EnvironmentSatisfaction. A large portion (about 25%) of ex-employees were dissatisfied with the environment. While the satisfaction was high among the retained employees.

3. Significant number (about 50%) of employees who left had lower JobLevel. Therefore, it seems that they left while looking for higher JobLevel somewhere else.

4. Roughly 50% of the people who left had the lowest MonthlyIncome. Since lower income corresponds to lower JobLevel, this observation is sensible and is in line with the above point.

5. Most of the people who left were offered lower stock options as indicated by StockOptionLevel. People on lower job level are entrusted with lower responsibilities, and therefore have lower stake in the company. 

6. Attrition rate was higher in employees who had worked for less than 5 years, as indicated by TotalWorkingYears. Entry-level employees can have aggressive career plans which makes them switch in hope of better opportunities.

7. It seems that in people who were staying with the company for less than 5 years, had higher attrition rate, as indicated by YearsAtCompany. With time employees become loyal to the company.

8. According to YearsInCurrentRole, if a person continues in the same role for about a year, it is likely that he will quit the company.

9. Most of the ex-employees had worked OverTime. Therefore, it seems that OverTime is an indicator of employee churn.

Conclusion: Assuming that the churning rate is 20%, most of the attrition is due to the entry-level employees who have lower income, job level, and responsibilities. They probably leave the company hoping to land in better careers. High attrition is also associated with over time. A fraction of the attrition could also be linked to lower environment satisfaction. Employee churn can be significantly reduced by giving incentives to the young employees, reducing over time, and by improving the work environment.

```python
print 'Assuming that the data was recorded over a year, the churn rate is:', ((df['Attrition'] == 'Yes').sum() / float((df['Attrition'] == 'No').sum()))*100, '%'
```

```python
for col in df.columns[0:-1]:
    if df[col].dtype == int:
        fig = plt.figure()
        _ = plt.hist([df[df.Attrition == 'Yes'][col], df[df.Attrition == 'No'][col]], label=['Yes','No'], bins=20, normed=True)
    else:
        continue
    plt.xlabel(col)
    plt.ylabel('Normalized Frequency')
    plt.legend()
```

```python
for col in df.columns[0:-1]:
    fig = plt.figure()
    if df[col].dtype == object:
        _ = sns.countplot(x=col, hue="Attrition", data=df)
    else:
        continue
```
